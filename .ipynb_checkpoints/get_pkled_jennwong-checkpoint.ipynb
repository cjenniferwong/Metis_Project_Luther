## Now that I have all the charts in pkl files
I want to merge them together into one large data frame so that I can count how many times each unique song will appear to be merged with the kaggle site

```python
from glob import glob
import pandas as pd
```

```python
file_paths = glob('*.pkl', recursive=True)

complete_df = pd.DataFrame()
list_of_df = []
for pkl in file_paths:
    list_of_df.append(pd.read_pickle(pkl))
    
complete_df = pd.concat(list_of_df)
complete_df.to_pickle('complete_df.pkl')
```

```python
complete_df.groupby(['Artist','Song']).count()
```
